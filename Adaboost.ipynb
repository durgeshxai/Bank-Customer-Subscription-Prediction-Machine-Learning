```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Copy the original data
x_encoded = x.copy()
y_encoded = y.copy()
```

```python
# Handle missing values in X_encoded
x_encoded = x_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
# Handle missing values in X_encoded
y_encoded = y_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
# Label Encoding for object columns
le_dict = {}
```

```python
for col in x_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    x_encoded[col] = le.fit_transform(x_encoded[col])
```

```python
x_encoded
```

```python
for col in y_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    y_encoded[col] = le.fit_transform(y_encoded[col])
    le_dict[col] = le
```

```python
y_encoded
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_estimator = DecisionTreeClassifier
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adaboost_model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=50)
```

```python
adaboost_model.fit(x_train, y_train)
```

```python
y_pred = adaboost_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
