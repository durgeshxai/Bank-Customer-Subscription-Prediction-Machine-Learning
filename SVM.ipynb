```python
from sklearn.preprocessing import LabelEncoder
```

```python
x_encoded = x.copy()
y_encoded = y.copy()
```

```python
# Handle missing values in X_encoded
x_encoded = x_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
# Handle missing values in X_encoded
y_encoded = y_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
le_dict = {}
```

```python
for col in x_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    x_encoded[col] = le.fit_transform(x_encoded[col])
```

```python
x_encoded
```

```python
for col in y_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    y_encoded[col] = le.fit_transform(y_encoded[col])
```

```python
y_encoded
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='rbf')
```

```python
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
