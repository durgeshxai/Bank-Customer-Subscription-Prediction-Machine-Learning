```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Copy the original data
x_encoded = x.copy()
y_encoded = y.copy()
```

```python
# Handle missing values in X_encoded
x_encoded = x_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
# Handle missing values in X_encoded
y_encoded = y_encoded.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(col.median()))
```

```python
# Label Encoding for object columns
le_dict = {}
```

```python
for col in x_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    x_encoded[col] = le.fit_transform(x_encoded[col])
```

```python
x_encoded
```

```python
for col in y_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    y_encoded[col] = le.fit_transform(y_encoded[col])
    le_dict[col] = le
```

```python
y_encoded
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2)
```

```python
# Convert y to Series if needed
if isinstance(y_train, pd.DataFrame):
    y_train = y_train.iloc[:, 0]
    y_test = y_test.iloc[:, 0]
```

```python
from xgboost import XGBClassifier
```

```python
# Train model
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(x_train, y_train)
```

```python
# Predict
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

# SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(sampling_strategy='minority') 
x,y=smote.fit_resample(x_encoded,y_encoded)
y.value_counts()
```

```python
x_encoded = x
y_encoded = y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2)
```

```python
# Train model
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
